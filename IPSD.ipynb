***1.Mengenali Karakteristik Data***

***
terdapat beberapa jenis karakteristik data diantaranya :

- Ukuran Data meliputi jumlah baris dan kolom
- tipe data tiap kolom meliputi numerik, kategorikal, teks
- ringkasan statistik meliputi mean, median, min, max, std
- Distribusi Data meliputi histogram, boxplot
- Missing Value (jumlah data kosong)
- Korelasi antar Variabel



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh data
data = {
    'Umur': [23, 45, 12, 36, 52, np.nan, 40],
    'Jenis_Kelamin': ['Pria', 'Wanita', 'Wanita', 'Pria', 'Pria', 'Wanita', 'Pria'],
    'Pendapatan': [5000, 7000, 2000, 6500, 8000, 7200, np.nan]
}

df = pd.DataFrame(data)

# 1. Ukuran Data (jumlah baris dan kolom)
print("Ukuran data:", df.shape)

# 2. Tipe data tiap kolom
print("\nTipe data per kolom:\n", df.dtypes)

# 3. Ringkasan statistik
print("\nRingkasan Statistik:\n", df.describe(include='all'))

# 4. Distribusi Data (Histogram & Boxplot)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(df['Umur'], bins=5, kde=True)
plt.title("Distribusi Umur")

plt.subplot(1,2,2)
sns.boxplot(y=df['Pendapatan'])
plt.title("Boxplot Pendapatan")
plt.show()

# 5. Missing Value
print("\nJumlah Missing Value:\n", df.isnull().sum())

# 6. Korelasi antar Variabel
print("\nKorelasi:\n", df.corr(numeric_only=True))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Heatmap Korelasi")
plt.show()
```

***5.Memanipulasi Matriks dan Operasi operasi Matriks***

#### Manipulasi Matriks berarti : mengubah, mengambil, atau mengatur bentuk/ isi Matriks (indexing & slicing, reshape, concat, transpose)
#### Operasi Matriks berarti : perhitungan Matematis (Penjumlahan, Perkalian, dll)

```python
#contoh manipulasi matriks :
import numpy as np

# Membuat matriks A
A = np.array([[1, 2, 3],
              [4, 5, 6]])
print("Matriks A:\n", A)

# Indexing
print("Elemen baris 0 kolom 1:", A[0, 1])   # ambil angka 2
print("Baris kedua:", A[1, :])             # ambil baris [4 5 6]
print("Kolom pertama:", A[:, 0])           # ambil kolom [1 4]

# Reshape
B = np.arange(6).reshape(2, 3)
print("Matriks B:\n", B)

# Concatenate (gabung vertikal dan horizontal)
print("Gabung vertikal A dengan A:\n", np.concatenate((A, A), axis=0))
print("Gabung horizontal A dengan A:\n", np.concatenate((A, A), axis=1))

# Transpose
print("Transpose A:\n", A.T)
```

```python
# contoh operasi matriks
import numpy as np

# Matriks contoh
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("Matriks A:\n", A)
print("Matriks B:\n", B)

# 1. Penjumlahan & Pengurangan
print("A + B:\n", A + B)
print("A - B:\n", A - B)

# 2. Perkalian elemen-per-elemen
print("A * B:\n", A * B)

# 3. Perkalian matriks (dot product)
print("A dot B:\n", np.dot(A, B))   # atau A @ B

# 4. Determinan & Invers A
det_A = np.linalg.det(A)
inv_A = np.linalg.inv(A)

print("Determinan A:", det_A)
print("Invers A:\n", inv_A)

# 5. Nilai Eigen & Vektor Eigen
vals, vecs = np.linalg.eig(A)
print("Nilai Eigen A:", vals)
print("Vektor Eigen A:\n", vecs)
```

***7. Mengetahui indeks elemen pada array***

```python
# langung dengan indeks
import numpy as np

A = np.array([[10, 20, 30],
              [40, 50, 60]])

print(A[0, 1])   # elemen baris 0 kolom 1  20
print(A[1, 2])   # elemen baris 1 kolom 2  60
```

***8.Mengimplementasikan operasi logika matematika menggunakan Boolean***

 Array NumPy bisa dibandingkan dengan nilai tertentu atau array lain.
Hasil perbandingan berupa array Boolean (True / False).
Boolean ini bisa digunakan untuk filtering data, atau sebagai dasar operasi logika.


```python
#Contoh
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

# Operasi logika
print(arr > 3)        # [False False False  True  True]
print(arr % 2 == 0)   # [False  True False  True False]

# Filtering dengan boolean
print(arr[arr > 3])   # [4 5]
print(arr[arr % 2 == 0]) #[24]
```

***9. Melakukan operasi aritmatika***

Operasi aritmatika pada NumPy dilakukan per elemen (vectorized).
Mendukung penjumlahan, pengurangan, perkalian, pembagian, pangkat, dll.
Operasi aritmatika pada NumPy dilakukan per elemen (vectorized).


```python
#Contoh
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Penjumlahan:", a + b)  # [5 7 9]
print("Pengurangan:", a - b)  # [-3 -3 -3]
print("Perkalian:", a * b)    # [ 4 10 18]
print("Pembagian:", a / b)    # [0.25 0.4 0.5]
print("Pangkat:", a ** 2)     #[149]
```

***10. Elementwise function***

Fungsi matematika bisa diterapkan ke tiap elemen array secara otomatis.
Tidak perlu perulangan manual.
Contohnya: akar, sinus, logaritma, eksponensial.


```python
import numpy as np
arr = np.array([1, 4, 9, 16])

print("Akar:", np.sqrt(arr))   # [1. 2. 3. 4.]
print("Eksponensial:", np.exp(arr))  # [2.7e+00 ... 8.9e+06]
print("Sinus:", np.sin(arr))   # [-0.84 -0.76  0.41 -0.29]
print("Logaritma:", np.log(arr))  # [0.1.382.192.77]
```

***12.Melakukan Operasi dengan Logika Himpunan***
Array di NumPy bisa diperlakukan seperti himpunan dalam matematika.
Operasi yang umum:

Union = gabungan dua array tanpa duplikat

Intersection = irisan (elemen yang sama)

Difference = selisih (elemen yang ada di satu array tapi tidak di yang lain)

Unique = ambil elemen unik dari array

```python
# Contoh:

a = np.array([1, 2, 3, 4, 5])
b = np.array([3, 4, 5, 6, 7])

print("Union (gabungan):", np.union1d(a, b))
print("Intersection (irisan):", np.intersect1d(a, b))
print("Difference (a-b):", np.setdiff1d(a, b))
print("Unique (tanpa duplikat):", np.unique([1,2,2,3,3,4]))
```

***13.Memanipulasi Array***

Manipulasi array meliputi mengubah bentuk, memotong, menggabungkan, atau memisahkan array.
Beberapa contoh operasi umum:

Reshape = ubah bentuk array (misalnya 1D jadi 2D)

Transpose = tukar baris dan kolom

Indexing & Slicing = ambil elemen tertentu

Concatenate = gabungkan array

Split = pecah array jadi beberapa bagian

```python
#Contoh:

# Membuat array 1-9
c = np.arange(1, 10)

# Reshape ke 3x3
reshaped = c.reshape(3, 3)
print("Reshape 3x3:\n", reshaped)

# Transpose
print("Transpose:\n", reshaped.T)

# Indexing & Slicing
print("Ambil elemen pertama:", c[0])
print("Ambil 3 elemen pertama:", c[:3])

# Concatenate (gabung array)
d = np.array([10, 11, 12])
print("Concatenate:", np.concatenate([c, d]))

# Split (bagi array jadi 3 bagian)
split_arr = np.array_split(c, 3)
print("Split jadi 3:", split_arr)
```
