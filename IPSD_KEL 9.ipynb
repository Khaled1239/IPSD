```python
#Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#####################################################################################
# 1. Load Data
df = pd.read_csv("/content/IPSD.csv")

#####################################################################################
# 2. Show Info
df.info()

#####################################################################################
# 3. Drop Duplicates & Nilai Unik
df.drop_duplicates(inplace=True)

for i in df.columns:
  print(f"Kolom : {i}")
  print(f"Unik : {df[i].unique()}")
  print(f"Jumlah : {df[i].nunique()}")

####################################################################################
# 4. Visualisasi Unik

Total_Unique = df.nunique()

plt.figure(figsize=(16, 9))
Total_Unique.plot(kind='bar', color='skyblue')
plt.title('Jumlah Nilai Unik Setiap Kolom')
plt.xlabel('Kolom')
plt.ylabel('Jumlah Nilai Unik')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. Menemukan Null Values

df.isnull().sum()
```

```python
# 6. Mengisi Null Values

df['Study Environment'].fillna(df['Study Environment'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
# 7. Melihat Tipe Data

df.dtypes
```

```python
# 8. Filter Data

dFiltered = df[df['Study Environment'] == 'Peaceful']
dFiltered.head()
```

```python
# 9. Boxplot Kolom Numeric

plt.figure(figsize=(16,9))
sns.boxplot(data=df.select_dtypes(include=["number"]))
plt.title("Box Plot untuk Kolom Numerik", fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# 10. Visualisasi korelasi dengan heatmap

corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```
